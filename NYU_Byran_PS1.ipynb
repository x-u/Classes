```python
import pandas as pd
from pandas import DataFrame
import numpy as np
from scipy.optimize import minimize
pd.options.display.max_colwidth=500
pd.set_option('display.precision',4)
```

```python
df = pd.read_stata('/Users/Titanic/Dropbox/3Classes/Bryan/PS1/YogurtLong.dta')
df.head()
```

```python
B = df.ix[:,3:8].values
P = df.ix[:,8:12].values
F = df.ix[:,12:16].values
```

```python
def Li(x):
    a = x[0:4] # take first 4 values as alpha
    b1 = x[4] # coeff on prices
    b2 = x[5] # coeff on features
    E = exp(a + b1*P + b2*F) # build matrix (N*J) for each exp
    E = np.insert(E, 0, 1, axis=1) # faster this way than to add 0 column first and then calculate exp(0)=1
    denom = E.sum(1)
    return -log((B * E).sum(1) / denom).sum()
```

```python
x0 = np.array([.22,.38, -2.2, -.18,
              -8.4, .015])
minimize(Li, x0, method='Nelder-Mead',options={'xtol': 1e-8, 'disp': True, 'maxfev':5000})
```
